```python
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam giác không hợp lệ")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Diện tích =",dt)
```

```python
from random import randint
while True:
    somay = randint(1, 101)
    solandoan = 0
    win = False
    while solandoan < 7:
        solandoan += 1
        songuoi = int(input("Nhập số bạn đoán [1...100]: "))
        print("Lần đoán thứ", solandoan)
        if somay == songuoi:
            print("Chúc mừng bạn đã đoán đúng, số của máy là", somay)
            win = True
            break
        if somay > songuoi:
            print("Bạn đoán sai, Số của máy > số bạn đoán")
        else:
            print("Bạn đoán sai, Số của máy < số bạn đoán")
    if not win:
        print("Bạn đã thua, số của máy là", somay)
    hoi = input("Bạn có muốn chơi tiếp không (c/k)? ")
    if hoi=="k":
        break
    print("Cám ơn bạn đã chơi Game!")
```

```python
def BMI (height, weight):
    return weight / (height **2)
def PhanLoai(bmi):
    if bmi < 18.5:
        return "Gầy"
    elif bmi <= 24.9:
        return "Bình thường"
    elif bmi <= 29.9:
        return "Thừa cân"
    elif bmi <= 34.9:
        return "Béo phì cấp độ 1"
    elif bmi <= 39.9:
        return "Béo phì cấp độ 2"
    else:
        return "Béo phì cấp độ 3"
def NguyCoBenh(bmi):
    if bmi < 18.5:
        return "Nguy cơ bệnh thấp"
    elif bmi <= 24.9:
        return "Nguy cơ bệnh trung bình"
    elif bmi <= 29.9:
        return "Nguy cơ bệnh cao"
    elif bmi <= 34.9:
        return "Nguy cơ bệnh cao"
    elif bmi <= 39.9:
        return "Nguy cơ bệnh rất cao"
    else:
        return "Nguy cơ bệnh nguy hiểm"
print("Chương trình tính chỉ số BMI")
height = float(input("Nhập chiều cao (mét): "))
weight = float(input("Nhập cân nặng (kg): "))
bmi = BMI(height, weight)
print("Chỉ số BMI của bạn là: ", BMI(height, weight))
print("Phân loại: ", PhanLoai(bmi))
print("Nguy cơ bệnh của Thím! là: ", NguyCoBenh(bmi))
```
