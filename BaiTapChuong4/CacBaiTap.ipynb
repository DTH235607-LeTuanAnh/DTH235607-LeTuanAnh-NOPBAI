```python
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam giác không hợp lệ")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Diện tích =",dt)
```

```python
from random import randint
while True:
    somay = randint(1, 101)
    solandoan = 0
    win = False
    while solandoan < 7:
        solandoan += 1
        songuoi = int(input("Nhập số bạn đoán [1...100]: "))
        print("Lần đoán thứ", solandoan)
        if somay == songuoi:
            print("Chúc mừng bạn đã đoán đúng, số của máy là", somay)
            win = True
            break
        if somay > songuoi:
            print("Bạn đoán sai, Số của máy > số bạn đoán")
        else:
            print("Bạn đoán sai, Số của máy < số bạn đoán")
    if not win:
        print("Bạn đã thua, số của máy là", somay)
    hoi = input("Bạn có muốn chơi tiếp không (c/k)? ")
    if hoi=="k":
        break
    print("Cám ơn bạn đã chơi Game!")
```

```python
def BMI (height, weight):
    return weight / (height **2)
def PhanLoai(bmi):
    if bmi < 18.5:
        return "Gầy"
    elif bmi <= 24.9:
        return "Bình thường"
    elif bmi <= 29.9:
        return "Thừa cân"
    elif bmi <= 34.9:
        return "Béo phì cấp độ 1"
    elif bmi <= 39.9:
        return "Béo phì cấp độ 2"
    else:
        return "Béo phì cấp độ 3"
def NguyCoBenh(bmi):
    if bmi < 18.5:
        return "Nguy cơ bệnh thấp"
    elif bmi <= 24.9:
        return "Nguy cơ bệnh trung bình"
    elif bmi <= 29.9:
        return "Nguy cơ bệnh cao"
    elif bmi <= 34.9:
        return "Nguy cơ bệnh cao"
    elif bmi <= 39.9:
        return "Nguy cơ bệnh rất cao"
    else:
        return "Nguy cơ bệnh nguy hiểm"
print("Chương trình tính chỉ số BMI")
height = float(input("Nhập chiều cao (mét): "))
weight = float(input("Nhập cân nặng (kg): "))
bmi = BMI(height, weight)
print("Chỉ số BMI của bạn là: ", BMI(height, weight))
print("Phân loại: ", PhanLoai(bmi))
print("Nguy cơ bệnh của Thím! là: ", NguyCoBenh(bmi))
```

```python
def ROI(dt,cp):
     return (dt-cp)/cp
def GoiYDauTu(roi):
 if roi>=0.75:
     return "Nên đầu tư"
 else:
     return "Không nên đầu tư"
print("Chương trình tính ROI")
dt=int(input("Nhập Doanh Thu:"))
cp=int(input("Nhập chi phí:"))
roi=ROI(dt,cp)
print("Tỉ Lệ ROI=",roi)
print("==>",GoiYDauTu(roi))
```

Trả lời: 

Hàm randrange(0, 100) trong Python sẽ sinh số nguyên ngẫu nhiên trong khoảng [0, 100), tức là:
- Bắt đầu từ 0
- Lớn nhất là 99
- Chỉ nhận giá trị nguyên (int), không bao gồm số thực

==> Các giá trị có thể xuất hiện khi chạy randrange(0, 100) là: 34, 0, 99

```python
import random

# Danh sách giá trị cần kiểm tra
values = [4.5, 34, -1, 100, 0, 99]

print("Kiểm tra các giá trị có thể xuất hiện trong randrange(0, 100):\n")
for v in values:
    if isinstance(v, int) and 0 <= v < 100:
        print(f"{v}: ✅ Có thể xuất hiện")
    else:
        print(f"{v}: ❌ Không thể xuất hiện")
"""Hàm randrange(0, 100) trong Python sẽ sinh số nguyên ngẫu nhiên trong khoảng [0, 100), tức là:
Bắt đầu từ 0
Lớn nhất là 99
Chỉ nhận giá trị nguyên (int), không bao gồm số thực"""
```

```python
import math

"""
Chương trình tính độ dài đoạn thẳng AB
Nhập tọa độ điểm A(xA, yA) và B(xB, yB).
Công thức: d = sqrt((xB - xA)^2 + (yB - yA)^2)
"""

# Nhập tọa độ điểm A và B
xA = float(input("Nhập hoành độ xA: "))
yA = float(input("Nhập tung độ yA: "))
xB = float(input("Nhập hoành độ xB: "))
yB = float(input("Nhập tung độ yB: "))

# Tính độ dài đoạn AB
d = math.sqrt((xB - xA)**2 + (yB - yA)**2)

# Xuất kết quả
print(f"Độ dài đoạn AB = {d}")
```

```python
import math

print("Chương trình tính logₐx (logarit cơ số a của x)")
x = float(input("Nhập x (>0): "))
a = float(input("Nhập a (>0, a ≠ 1): "))

if x > 0 and a > 0 and a != 1:
    logax = math.log(x) / math.log(a)
    print(f"logₐx = {logax}")
else:
    print("Giá trị nhập không hợp lệ (x > 0, a > 0, a ≠ 1)")
```

```python
import math

print("Chương trình tính S(n) = √2 + √2 + ... + √2 (n dấu căn lồng nhau)")
n = int(input("Nhập n: "))

s = 0
for i in range(n):
    s = math.sqrt(2 + s)
print(f"S({n}) = {s}")
```
