```python
def CheckDoiXung(s):
    flag = True
    for i in range(len(s)):
        if s[i] != s[len(s) - i - 1]:
            flag = False
            break
    return flag
def main():
    s = input("Nhập vào 1 chuỗi: ")
    if CheckDoiXung(s):
        print("Chuỗi đối xứng")
    else:
        print("Chuỗi không đối xứng")
while True:
    main()
    s = input("Bạn có muốn tiếp tục không (c/k): ")
    if s == 'k':
        break
print("Chương trình kết thúc")
```

```python
def ToiUuChuoi(s):
    s2=s
    s2=s2.strip()
    arr=s2.split(' ')
    s2=""
    for x in arr:
        word=x
        if len(word.strip())!=0:
            s2+=word+" "
    return s2.strip()

s ="    Tôi   là     Lê     Tuấn      Anh   "
print(s,"=>",len(s))
s= ToiUuChuoi(s)
print(s,"=>",len(s))    
```
